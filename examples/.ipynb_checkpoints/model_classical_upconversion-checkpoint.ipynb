Three-wave mixing: $\omega_1+\omega_2=\omega_3$

The solution for the sum frequency wave is,
\begin{align}
a_3(L,t'_3)=-j\kappa_3\int_0^L dz a_1(t'_3-\eta_{13}z)a_2(t'_3-\eta_{23}z)
\end{align}

where

\begin{align}
\eta_{ij}=\frac{1}{v_{g,i}}-\frac{1}{v_{g,j}}
\end{align}

```python
#Crystal parameters
L=1e3
cr=crystal.BiBO()
angle=151.3
extra_ordinary=[angle,cr.ny,cr.nz] #Extra-ordinary
ordinary=cr.nx #Ordinary 

lambda_i0,sigma_li=0.82251,0.00409 #idler
lambda_s0,sigma_ls=0.7324,0.00353   #signal
lambda_g0,sigma_lg=0.7324,0.00177 #

wi0,sigma_i=w2f(lambda_i0,dx=sigma_li)
ws0,sigma_s=w2f(lambda_s0,dx=sigma_ls)
wg0,sigma_g=w2f(lambda_g0,dx=sigma_lg)

w30,w40=wg0+wi0,wg0+ws0

lambda_30,lambda_40=w2f(w30),w2f(w40)

dki,d2ki=1.0/GV(lambda_i0,cr.n_e,extra_ordinary),GVD(lambda_i0,cr.n_e,extra_ordinary) #idler
dks,d2ks=1.0/GV(lambda_s0,cr.n_e,extra_ordinary),GVD(lambda_s0,cr.n_e,extra_ordinary) #idler
dkg,d2kg=1.0/GV(lambda_g0,cr.n_e,extra_ordinary),GVD(lambda_g0,cr.n_e,extra_ordinary) #signal
        
#GV and GVD
dk3,d2k3=1.0/GV(lambda_30,cr.sellemeir,ordinary),GVD(lambda_30,cr.sellemeir,ordinary) #pump
dk4,d2k4=1.0/GV(lambda_40,cr.sellemeir,ordinary),GVD(lambda_40,cr.sellemeir,ordinary)

print(dki-dk3)
print(dkg-dk3)
print(dks-dk3)
```

## Idler

```python
A=50000
lambda_i0,sigma_li=0.82251,0.00409 #idler
lambda_g0,sigma_lg=0.775,0.00177 #

wi0,sigma_i=w2f(lambda_i0,dx=sigma_li)
wg0,sigma_g=w2f(lambda_g0,dx=sigma_lg)

sigma_i=0.5/120
sigma_g=0.5/120

tau,I3=model_gaussian_sfg(wi0,sigma_i,wg0,sigma_g,A=A)
```

```python
fit=fit_gaussian((tau,I3)).x
plt.plot(tau,I3,'.C1')
plt.plot(tau,gauss(tau,*fit),'C1')
plt.title('Optical gating the idler')
plt.xlabel(r'$\tau$')
plt.show()
print('Measured width',fit[2])
print('Expected width',np.sqrt(0.25/sigma_g**2+0.25/sigma_i**2+4*A**2*sigma_i**2))
```

## Signal

```python
A=50000
lambda_s0,sigma_ls=0.7324,0.00353   #signal
lambda_g0,sigma_lg=0.7324,0.00177 #

ws0,sigma_s=w2f(lambda_s0,dx=sigma_ls)
wg0,sigma_g=w2f(lambda_g0,dx=sigma_lg)

sigma_s=0.5/120
sigma_g=0.5/120

tau,I3=model_gaussian_sfg(ws0,sigma_s,wg0,sigma_g,A=A)
```

```python
fit=fit_gaussian((tau,I3)).x
plt.plot(tau,I3,'.C1')
plt.plot(tau,gauss(tau,*fit),'C1')
plt.title('Optical gating the signal')
plt.xlabel(r'$\tau$')
plt.show()
print('Measured width',fit[2])
print('Expected width',np.sqrt(0.25/sigma_g**2+0.25/sigma_s**2+4*A**2*sigma_s**2))
```

## Trends

```python
lambda_i0,sigma_li=0.82251,0.00409 #idler
lambda_g0,sigma_lg=0.77551,0.00177 #

wi0,sigma_i=w2f(lambda_i0,dx=sigma_li)
wg0,sigma_g=w2f(lambda_g0,dx=sigma_lg)

sigma_g=0.5/120
sigma_i=0.5/120
A=np.arange(0,5e4,5e3)
mwidth=np.zeros(A.size)
ewidth=np.zeros(A.size)

for i in range(A.size):
    tau,I3=model_gaussian_sfg(wi0,sigma_i,wg0,sigma_g,A=A[i],dt=10,L=1000)
    mwidth[i]=fit_gaussian((tau,I3)).x[2]
    ewidth[i]=np.sqrt(0.25/sigma_g**2+0.25/sigma_i**2+4*A[i]**2*sigma_i**2)
    
```

```python
plt.figure()
plt.title('L=%i um'%1000)
plt.plot(A,mwidth,label='With phase-matching')
plt.plot(A,ewidth,label='No phase-matching')
plt.xlabel(r'Chirp $(fs^2)$')
plt.ylabel(r'RMS width (fs)')
plt.legend()
plt.show()
```
